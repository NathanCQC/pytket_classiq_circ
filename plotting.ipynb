```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files into pandas dataframes
df_classiq = pd.read_csv('tket_classiq_qsvt.csv', index_col=0)
df_nathan = pd.read_csv('tket_nathan_qsvt.csv', index_col=0)

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot time
axs[0, 0].plot(df_classiq.columns, df_classiq.loc['time'], label='Classiq_pytket', marker='o')
axs[0, 0].plot(df_nathan.columns, df_nathan.loc['time'], label='Nathan', marker='o')
axs[0, 0].set_title('Time')
axs[0, 0].set_xlabel('Size')
axs[0, 0].set_ylabel('Time')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot depth
axs[0, 1].plot(df_classiq.columns, df_classiq.loc['depth'], label='Classiq_pytket', marker='x')
axs[0, 1].plot(df_nathan.columns, df_nathan.loc['depth'], label='Nathan_pyktet', marker='x')
axs[0, 1].set_title('Depth')
axs[0, 1].set_xlabel('Size')
axs[0, 1].set_ylabel('Depth')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot cx_counts
axs[1, 0].plot(df_classiq.columns, df_classiq.loc['cx_counts'], label='Classiq_pytket', marker='s')
axs[1, 0].plot(df_nathan.columns, df_nathan.loc['cx_counts'], label='Nathan_pyktet', marker='s')
axs[1, 0].set_title('CX Counts')
axs[1, 0].set_xlabel('Size')
axs[1, 0].set_ylabel('CX Counts')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot width
axs[1, 1].plot(df_classiq.columns, df_classiq.loc['width'], label='Classiq_pytket', marker='d')
axs[1, 1].plot(df_nathan.columns, df_nathan.loc['width'], label='Nathan_pyktet', marker='d')
axs[1, 1].set_title('Width')
axs[1, 1].set_xlabel('Size')
axs[1, 1].set_ylabel('Width')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
